##### Scraping artifacts

```python
import pandas as pd

listings = pd.read_csv("LosAngelesCaliforniaUnitedStates/listings/listings.csv", low_memory=False)

# How many listings were freshly found vs. just carried over from a prior scrape?
print("Source breakdown:")
print(listings['source'].value_counts())
print()
print("As percentage:")
print((listings['source'].value_counts(normalize=True) * 100).round(1))

# Check the scrape date itself — is this really "current" data?
print("Unique scrape dates:", listings['last_scraped'].unique())

# Check how many decimal places coordinates actually have —
# real GPS is usually 6+ decimals; heavy rounding suggests jittering/anonymization
sample_coords = listings[['latitude', 'longitude']].head(10)
print(sample_coords)

# Count how many *unique* lat/long pairs exist vs. total listings —
# if many listings share the exact same coordinate, that's a sign of deliberate snapping/jittering
print("\nTotal listings:", len(listings))
print("Unique coordinate pairs:", listings[['latitude','longitude']].drop_duplicates().shape[0]) 
```
